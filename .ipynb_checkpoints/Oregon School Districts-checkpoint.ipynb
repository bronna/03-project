```python
import pandas as pd
```

```python
import numpy as np
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
df = pd.read_csv("RCmediaDistrictsDisaggregated.csv")
```

```python
df = df[['District Name', 'County', 'Student Enrollment', 'White Students', 'White Teachers', 'District On-Time Graduation', 'Administrators', 'Teacher Experience', 'Counselors/Psychologists', 'Licensed Librarians', 'Teachers', 'Black/African American On-Time Graduation', 'Hispanic/Latino On-Time Graduation', 'White On-Time Graduation', 'Economically Disadvantaged On-Time Graduation', 'Students with Disabilities On-Time Graduation']]
```

```python
df = df.replace('*', np.nan)
```

```python
df.head()
```

```python
df.info(verbose=True)
```

```python
#removed commas in numbers
df["Student Enrollment"] = df["Student Enrollment"].str.replace(',', '')
```

```python
#removed commas in numbers
df["Teachers"] = df["Teachers"].str.replace(',', '')
```

```python
#changed any <1 in columns to 1
df['Counselors/Psychologists'] = df['Counselors/Psychologists'].str.replace('<1','1')
```

```python
#changed any <1 in columns to 1
df['Licensed Librarians'] = df['Licensed Librarians'].str.replace('<','')
```

```python
#changed any <1 in columns to 1
df['Administrators'] = df['Administrators'].str.replace('<','')
```

```python
#removed %
df['White Students'] = df['White Students'].str.replace('%','').replace('<','')
```

```python
#removed %
df['White Teachers'] = df['White Teachers'].str.replace('%','')
```

```python
#removed %
df['Teacher Experience'] = df['Teacher Experience'].str.replace('%','')
```

```python
df[['Student Enrollment', 'Teachers', 'Counselors/Psychologists', 'Licensed Librarians', 'Administrators', 'White Students', 'White Teachers', 'Teacher Experience']] = \
        df[['Student Enrollment', 'Teachers', 'Counselors/Psychologists', 'Licensed Librarians', 'Administrators', 'White Students', 'White Teachers', 'Teacher Experience']].astype(int)
```

```python
#removed any >
df["Economically Disadvantaged On-Time Graduation"] = df["Economically Disadvantaged On-Time Graduation"].str.replace('>','')
```

```python
#removed any %
df["Economically Disadvantaged On-Time Graduation"] = df["Economically Disadvantaged On-Time Graduation"].str.replace('%','')
```

```python
#removed any >
df["District On-Time Graduation"] = df["District On-Time Graduation"].str.replace('>','')
```

```python
#removed any %
df["District On-Time Graduation"] = df["District On-Time Graduation"].str.replace('%','')
```

```python
#removed any >
df["Black/African American On-Time Graduation"] = df["Black/African American On-Time Graduation"].str.replace('>','')
```

```python
#removed any %
df["Black/African American On-Time Graduation"] = df["Black/African American On-Time Graduation"].str.replace('%','')
```

```python
#removed any >
df["Hispanic/Latino On-Time Graduation"] = df["Hispanic/Latino On-Time Graduation"].str.replace('>','')
```

```python
#removed any >
df["Hispanic/Latino On-Time Graduation"] = df["Hispanic/Latino On-Time Graduation"].str.replace('%','')
```

```python
#removed any >
df["White On-Time Graduation"] = df["White On-Time Graduation"].str.replace('>','')
```

```python
#removed any >
df["White On-Time Graduation"] = df["White On-Time Graduation"].str.replace('%','')
```

```python
#removed any >
df["Students with Disabilities On-Time Graduation"] = df["Students with Disabilities On-Time Graduation"].str.replace('>','')
```

```python
#removed any >
df["Students with Disabilities On-Time Graduation"] = df["Students with Disabilities On-Time Graduation"].str.replace('%','')
```

```python
df[["Economically Disadvantaged On-Time Graduation", "District On-Time Graduation", "Black/African American On-Time Graduation", "Hispanic/Latino On-Time Graduation", "White On-Time Graduation", "Students with Disabilities On-Time Graduation"]] = \
        df[["Economically Disadvantaged On-Time Graduation", "District On-Time Graduation", "Black/African American On-Time Graduation", "Hispanic/Latino On-Time Graduation", "White On-Time Graduation", "Students with Disabilities On-Time Graduation"]].astype(float)
```

```python
df['librarian_student_ratio'] = df["Licensed Librarians"] / df["Student Enrollment"]
```

```python
df['counselor_student_ratio'] = df["Counselors/Psychologists"] / df["Student Enrollment"]
```

```python
df['teacher_student_ratio'] = df["Teachers"] / df["Student Enrollment"]
```

```python
df['administrator_student_ratio'] = df["Administrators"] / df["Student Enrollment"]
```

```python
df['Students_of_Color'] = 100 - df['White Students']
```

```python
df['Teachers_of_Color'] = 100 - df['White Teachers']
```
